# Plotting libraries

```julia
import Pkg 

Pkg.activate(temp=true) 

Pkg.add([
        Pkg.PackageSpec(name="ECharts", version="0.6"),
        Pkg.PackageSpec(name="GR", version="0.59"),
        Pkg.PackageSpec(name="Gadfly", version="1"),
        Pkg.PackageSpec(name="Gaston", version="1"),
        Pkg.PackageSpec(name="Gnuplot", version="1"),
        Pkg.PackageSpec(name="GracePlot", version="0.3"),
        Pkg.PackageSpec(name="InspectDR", version="0.4"),
        Pkg.PackageSpec(name="PGFPlotsX", version="1"),
        Pkg.PackageSpec(name="PlotlyJS", version="0.18"),
        Pkg.PackageSpec(name="UnicodePlots", version="2"),
        Pkg.PackageSpec(name="VegaLite", version="2"),
        Pkg.PackageSpec(name="Winston", version="0.15"),
        Pkg.PackageSpec(name="PyPlot", version="2")
        ]) 
```

**Note** that we will import the different plotting packages, and therefore, we will use the qualified names for their function. Most packages export the same names, so using qualified names resolves the ambiguities. Also, the use of `import` instead of `using` keeps clean the global namespace.

## Example data

Input data for the plot examples

```julia
x = range(0, 2pi, length=100)
```

```julia
y = sin.(x)
```

## UnicodePlots

```julia
import UnicodePlots
```

```julia
UnicodePlots.lineplot(x, y)
```

## Gadfly

```julia
import Gadfly
```

```julia
import Gadfly: Geom 
```

```julia
Gadfly.plot(x=x, y=y, Geom.line)
```

## Winston

```julia
import Winston
```

```julia
Winston.plot(x, y, "-")
```

## PyPlot

```julia
import PyPlot
```

```julia
PyPlot.plot(x, y)
```

## PlotlyJS

```julia
import PlotlyJS
```

```julia
PlotlyJS.plot(PlotlyJS.scatter(x=x, y=y, mode="lines"))
```

## GR

```julia
import GR
```

```julia
GR.plot(x, y)
```

## PGFPlotsX

```julia
import PGFPlotsX
```

```julia
PGFPlotsX.@pgf PGFPlotsX.Plot({ no_marks }, PGFPlotsX.Table(x, y))
```

## Gaston

```julia
import Gaston
```

```julia
Gaston.plot(x, y, w="l")
```

## Gnuplot

```julia
import Gnuplot
```

```julia
Gnuplot.@gp x y "with lines"
```

## GracePlot

```julia
import GracePlot
```

```julia
begin
	plt = GracePlot.new(fixedcanvas=false, emptyplot=false)
	g = GracePlot.graph(plt, 0)
	GracePlot.add(g, collect(x), y)
	GracePlot.autofit(g)
	GracePlot.redraw(plt)
end
```

## InspectDR

```julia
import InspectDR
```

```julia
begin
	figure = InspectDR.Plot2D()
	InspectDR.add(figure, collect(x), y)
	figure
end
```

## ECharts

```julia
import ECharts
```

```julia
ECharts.line(x, y)
```

## VegaLite

```julia
import VegaLite
```

```julia
VegaLite.@vlplot(:line, x=x, y=y)
```
