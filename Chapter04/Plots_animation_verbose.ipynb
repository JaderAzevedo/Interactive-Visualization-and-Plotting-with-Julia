```julia
using Plots
```

```julia
function plot_frame(angle)
    scatter(
        [cos(angle)], 
        [sin(angle)],
        ratio=:equal,
        xlims=(-1.5, 1.5), ylims=(-1.5, 1.5),
        legend=:none)
end
```

```julia
anim = Animation()
```

```julia
for angle in 0:0.05:pi
    plt = plot_frame(angle)
    frame(anim, plt)
end
```
